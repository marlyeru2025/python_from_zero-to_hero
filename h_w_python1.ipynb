- Write a program that takes a list of numbers from the user (until they enter ‘done’), then prints the list, its sorted version (ascending and descending)

```python
numbers = []
while True:
    user_input = input("Enter a number (or 'done' to finish): ")
    if user_input == 'done':
          break
    number=int(user_input)
    numbers.append(number)

print("Original list:", numbers)
numbers.sort()
print("Sorted list:", numbers)
numbers.sort(reverse=True)
print("Reverse sorted list:", numbers)
```

- Create a function modify_list(lst) that receives a list, appends 100 to it, inserts 999 at index 2, removes the first occurrence of 50 (if exists), pops the last element, and returns the modified list.

```python
def modify_list(lst):
  lst.append(100)
  lst.insert(2, 999)
  if 50 in lst:
    lst.remove(50)
  lst.pop()
  return lst
mynumbers = [10 , 3, 159, 50]
print(modify_list(mynumbers))
```

- Write a program that asks the user for 5 student names and their grades, stores them in a list of lists (e.g. [[‘Ali’, 85], [‘Sara’, 92]]), then prints the student with the highest and lowest grade.

```python
students=[]
for i in range(5):
  name=input("Enter student name: ")
  grade=int(input("Enter student grade: "))
  students.append([name,grade])
top= max(students , key=lambda x: x[1])
low= min(students , key=lambda x: x[1])
print("The student with the highest grade is:", top)
print("The student with the lowest grade is:", low)
```

- Create a function modify_fruits(fruits) that takes a list of fruits, does the following in order:
-
Appends “kiwi”
-
Inserts “grape” at index 2
-
Changes the item at index 0 to
uppercase version
-
Removes the second item (by index) using pop()
-
Returns the modified list

```python
def modify_fruits(fruits):
   fruits.append("kiwi")
   fruits.insert(2,"grape")
   fruits[0]=fruits[0].upper()
   fruits.pop(1)
   return fruits
fruits = ["banana" , "apple" , "orange"]
print( "fruits are " , modify_fruits(fruits))
```

- Implement a program that creates a list of temperatures (you choose 10 values), then uses a for loop to count how many are above 30 and how many below 15, and prints both counts.

```python
temperature = [10, 30, 20, 40, 11, 15, 22, 36, 25, 9]
above30 = 0
below15 = 0

for i in temperature:
    if i > 30:
        above30 = above30 + 1
    elif i < 15:
        below15 = below15 + 1
print("Final result above 30:", above30)
print("Final result below 15:", below15)
```

- Create a function remove_all_occurrences(lst, value) that removes all occurrences of a given value from the list

```python
def remove_all_occurrences(lst, val):
    for i in range(lst.count(val)):
        lst.remove(val)
    return lst
numbers=[2,3,2, 4, 5]
print(remove_all_occurrences(numbers, 2))
```

```python
target=42
attemps=0
while True:
  guess=(int(input("Enter your guess: ")))
  attemps=attemps+1
  if guess<target:
    print("too low")
  elif guess>target:
    print("too high")
  else:
    print ("correct")
    break
```
